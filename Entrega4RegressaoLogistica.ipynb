```python
import pandas as pd
import numpy as np

td = pd.read_csv('Employee.csv')

td.columns
print(td.dtypes)

td.head()
```

```python
td.drop(['City', 'Education', 'JoiningYear', 'PaymentTier', 'ExperienceInCurrentDomain'], axis=1, inplace=True)
td.columns
```

```python
td.isnull().sum()

td.info()
```

```python
gen = pd.get_dummies(td['Gender'], drop_first=True)
print(gen)
evenB = pd.get_dummies(td['EverBenched'], drop_first=True)
print(evenB)

td.drop(['Gender', 'EverBenched'], axis=1, inplace=True)
```

```python
td = pd.concat([td, gen, evenB], axis = 1)
td.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(td.drop('Age',axis=1), td['Age'], test_size=0.40, random_state=101)
print(X_train)
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(solver='lbfgs',max_iter=1000)


logmodel.fit(X_train,y_train)

predictions = logmodel.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, predictions)
print(conf_mat)
```
